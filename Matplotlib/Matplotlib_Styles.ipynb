```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data1 = np.linspace(0,10,20)
```

```python
data1
```

```python
data2 = data1 ** 2
```

```python
plt.subplots()
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2, color="#1E0703", alpha=0.8)
my_axes.plot(data2, data1)
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1, data1+2, color="blue", linewidth=1.0)
new_axes.plot(data1, data1+3, color="green", linewidth=4)
new_axes.plot(data1, data1+4, color="yellow", linestyle="--")
new_axes.plot(data1, data1+5, color="#000000", linestyle="-.")
new_axes.plot(data1, data1+6, color="#1e1023", linestyle=":")
new_axes.plot(data1, data1+7, color="blue", linestyle="--", marker="+", markersize=10, markerfacecolor="red")
new_axes.plot(data1, data1+8, color="green", linestyle="--", marker="o", markersize=6)
plt.show()
```
